```python
import numpy as np, matplotlib.pyplot as plt, cmath, matplotlib.patches as patches
from tqdm.notebook import tqdm
# 초기값 설정
ep = 1e-8 ; ep_s = 200 ; img=[] ; i=0 ; j=0 ; data=[[0,0,1]] 
#함수 정의
def f(x) :
  return x**3 -1
def df(x) :
  return 3*x**2

def conv(x) :
  arr=[0,0,0]; col = 0
  for i in range(3):    
    arr[i] = abs(cmath.phase(x)-phase_arr[i])
  if arr[0]>arr[1] and arr[2]>arr[1]:    
    col = 1
  elif arr[0]>arr[2] :    
    col = 2
  return col

convert_arr = [(np.cos(2*np.pi/3)+1j*np.sin(2*np.pi/3)),
               (np.cos(4*np.pi/3)+1j*np.sin(4*np.pi/3)),
               (1+0j)]
phase_arr = [cmath.phase(convert_arr[0]),
             cmath.phase(convert_arr[1]),
             cmath.phase(convert_arr[2])]

def draw(count,total,index):
  y1=data[count][0]-data[count][2] ; y2= data[count][0]+data[count][2] 
  x1=data[count][1]-data[count][2] ; x2=data[count][1]+data[count][2]
  i=0 ; j=0
  for x in tqdm(np.linspace(y1,y2,ep_s),desc="x",leave=False): 
    for y in np.linspace(x1,x2,ep_s):
      z = complex(x,y)
      ####Newton's Method
      while(abs(f(z))>ep):
        z -= f(z)/df(z)
      ##########
        color = conv(z)
        if color == 0: 
          img[j][i] = [255,0,0]
        elif color == 1: 
          img[j][i] = [0,255,0]
        elif color == 2: 
          img[j][i] = [0,0,255]
      j+=1
    i+=1;  j=0
  ax = plt.subplot(1,total,index)
  plt.imshow(img,origin='lower',extent=[y1,y2,x1,x2])
 
  #다음 장면 Rectangle
  if count < total-1 :
    rect = patches.Rectangle((data[count+1][0]-data[count+1][2],  #x1
                              data[count+1][1]-data[count+1][2]), #y1
                             2*data[count+1][2],                  #폭
                             2*data[count+1][2],                  #높이
                             linewidth=3,                         #선 굵기
                             edgecolor='black',                   #선 색
                             facecolor='none')
    ax.add_patch(rect)
  plt.ylabel('Imaginary');  plt.xlabel('Real')
  #plt.title("zoom in x,y:[{:.6f}~{:.6f}][{:.6f}~{:.6f}]".format(x1,x2,y1,y2)) 
  plt.title("zoom X $10^{}$".format(count)) 

##빈 칸 생성
for x in np.linspace(0,1,ep_s):
  img.append([])
  for y in np.linspace(0,1,ep_s):
    img[i].append([0,0,0])
  i+=1
```

```python
#장면 별 중심점 및 범위 list 추가
data=[[0,0,1]] 
for i in range(5):  data.append([10*1e-11+1.68760006e-1,8*1e-11+5.87370004e-1, 0.1**(i+1)])

plt.figure(figsize=(6*len(data),8))
plt.suptitle("Fractal of $z^{3}-1$", fontsize=4*len(data))

#processing
for i in tqdm(range(len(data)),desc="processing"):
  draw(i,len(data),i+1)
```

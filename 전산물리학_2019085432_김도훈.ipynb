###pi

````
import numpy as np
N=100000
eps = 1e-3
sum=0.0
n=0
x_arr=[]
y_arr=[]
pi_list=[]
err_list=[]
while(n<N):
  x=(-1)**n/(2*n+1)
  sum+=x
  if abs(x)<eps: break
  if (n%100==0): print(n, sum*4)
  n+=1
  x_arr.append(n)
  y_arr.append(sum*4)
  err_list.append(abs(sum*4-np.pi))
  pi_list=np.pi*np.ones(len(x_arr))

plt.figure(figsize=(10,5))
plt.subplot(121) #1번째 줄을 2칸으로 나눈 뒤 1번째에 plot.
plt.plot(x_arr,y_arr, x_arr,pi_list)
plt.xscale('log')

plt.subplot(122) #1번째 줄을 2칸으로 나눈 뒤 2번째에 plot.
plt.plot(x_arr,err_list)

plt.ylabel('error',fontsize=15)
plt.xscale('log')
````

###리만

````
liman = lambda x: 1/(x**2)
n_list, y_list, err_list = [],[],[]
zeta_2=np.pi**2/6
sum=0.0
N=10000
n_list = list(range(1,N))
for x in range(1,N):
  sum+=liman(x)
  y_list.append(sum)
  err_list.append(abs(sum-zeta_2))
z_list=zeta_2*np.ones_like(y_list)
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(n_list,y_list,n_list,z_list)
plt.xscale('log')
plt.xlabel('Iteration',fontsize=15)
plt.subplot(122)
plt.plot(n_list,err_list)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Iteration',fontsize=15)
plt.ylabel('abs(Error)',fontsize=15)
plt.grid()
````
